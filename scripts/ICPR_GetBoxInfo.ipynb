```
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
from google.colab.patches import cv2_imshow #because cv2.imshow() causes colab kernel to crash
import os
```

```
f = open('/content/drive/My Drive/ICPR/segmented-images/bounding-boxes.json',)
boxes=json.load(f)
```

```
#Now that we have the json file, we will find out the image names
images=[]
for k in boxes.keys():
  images.append(k)

print(images)
print(len(images))
```

```
vals= boxes.values()
# print(vals)
```

```
len(vals)
```

```
xmin=[]
xmax=[]
ymin=[]
ymax=[]
height=[]
width=[]
```

```
#dividing them
for v in vals:
  s=str(v)
  x=s.split(',')
  height.append(x[0])
  width.append(x[1])
  xmin.append(x[3])
  ymin.append(x[4])
  xmax.append(x[5])
  ymax.append(x[6])
```

```
print(len(xmin))
print(len(ymin))
print(len(xmax))
print(len(ymax))
print(len(height))
print(len(width))
```

```
xmin1=[]
xmax1=[]
ymin1=[]
ymax1=[]
height1=[]
width1=[]
```

```
#characters to be removed
bad=['h','e','i','g','h','t','{',':',"'"," ",'x','m','n','a','y','w','d','}',']']
```

```
for ht in xmin:
  test_string=str(ht)
  for h in bad:
    test_string = test_string.replace(h, '')
  xmin1.append(int(test_string))
# xmin1
```

```
for ht in xmax:
  test_string=str(ht)
  for h in bad:
    test_string = test_string.replace(h, '')
  xmax1.append(int(test_string))
# xmax1
```

```
for ht in height:
  test_string=str(ht)
  for h in bad:
    test_string = test_string.replace(h, '')
  height1.append(int(test_string))
# height1
```

```
for ht in width:
  test_string=str(ht)
  for h in bad:
    test_string = test_string.replace(h, '')
  width1.append(int(test_string))
# width1
```

```
for ht in ymin:
  test_string=str(ht)
  for h in bad:
    test_string = test_string.replace(h, '')
  ymin1.append(int(test_string))
# ymin1
```

```
for ht in ymax:
  test_string=str(ht)
  for h in bad:
    test_string = test_string.replace(h, '')
  ymax1.append(int(test_string))
# ymax1
```

```
print(len(xmin1))
print(len(ymin1))
print(len(xmax1))
print(len(ymax1))
print(len(height1))
print(len(width1))
```

```
datafile=pd.DataFrame(images,columns=['image'])
datafile['xmin']=xmin1
datafile['xmax']=xmax1
datafile['ymin']=ymin1
datafile['ymax']=ymax1
datafile['height']=height1
datafile['width']=width1
```
